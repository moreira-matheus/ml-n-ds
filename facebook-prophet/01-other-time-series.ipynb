```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import entropy
from fbprophet import Prophet
%matplotlib inline
```

```python
def gen_column_mapper(columns):
    if len(columns) == 2:
        return dict(zip(columns, ['ds', 'y']))

    raise ValueError('Must have two columns!')
```

```python
kl_divergence = lambda p, q: entropy(pk=p, qk=q)
```

### Shampoo Sales Dataset

```python
shampoo = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/shampoo.csv')

shampoo.head()
```

```python
def parse_date(year_month, reference_year=1998):
    relative_year, month = year_month.split('-')
    actual_year = reference_year + int(relative_year) - 1
    return pd.to_datetime(str(actual_year) + '-' + month + '-01')
```

```python
shampoo['Month'] = shampoo.Month.apply(parse_date)
```

```python
series = shampoo.rename(columns=gen_column_mapper(shampoo.columns))
model_shampoo = Prophet()
model_shampoo.fit(series)
```

```python
future_shampoo = model_shampoo.make_future_dataframe(periods=12, freq='M')

future_shampoo.tail()
```

```python
forecast = model_shampoo.predict(future_shampoo)
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.set_title('Shampoo Sales')

model_shampoo.plot(forecast, ax=ax, xlabel='Month/Year', ylabel='Monthly sales')
ax.axvline(x=shampoo.iloc[-1]['Month'], color='darkblue', ls='dashed', lw='0.75', label='Last observation')

ax.legend()
plt.show()
```

```python
kl_divergence(p=shampoo.Sales, q=forecast.yhat.loc[shampoo.index])
```

### Minimum Daily Temperatures Dataset

```python
temperatures = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv')

temperatures.head()
```

```python
temperatures['Date'] = temperatures.Date.apply(pd.to_datetime)
```

```python
model_temp = Prophet()
model_temp.fit(temperatures.rename(columns=gen_column_mapper(temperatures.columns)))
```

```python
future_temp = model_temp.make_future_dataframe(periods=2*365, freq='D')

future_temp.tail()
```

```python
forecast_temp = model_temp.predict(future_temp)
```

```python
fig, ax = plt.subplots(figsize=(16,4))

ax.set_title('Minimum Daily Temperatures: Melbourne, AUS')

model_temp.plot(forecast_temp, ax=ax, xlabel='Day', ylabel='Min. Temperature')

ax.axvline(x=temperatures.iloc[-1]['Date'], color='darkblue', ls='dashed', lw='0.75', label='Last observation')

ax.legend()
plt.show()
```

```python
kl_divergence(p=temperatures.Temp,q=forecast_temp.yhat.loc[temperatures.index])
```

### Monthly Sunspot Dataset

```python
sunspots = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/monthly-sunspots.csv')

sunspots.head()
```

```python
sunspots['Month'] = sunspots.Month.apply(pd.to_datetime)
```

```python
model_sun = Prophet(seasonality_mode='multiplicative')
model_sun.fit(sunspots.rename(columns=gen_column_mapper(sunspots.columns)))
```

```python
future_sun = model_sun.make_future_dataframe(periods=30*12, freq='M')

future_sun.tail()
```

```python
forecast_sun = model_sun.predict(future_sun)
```

```python
fig, ax = plt.subplots(figsize=(16,6))

ax.set_title('Monthly Count of Observed Sunspots')

model_sun.plot(forecast_sun, ax=ax, xlabel='Month/Year', ylabel='Number of sunspots')

ax.axvline(x=sunspots.iloc[-1]['Month'], color='darkblue', ls='dashed', lw='0.75', label='Last observation')

ax.legend()
plt.show()
```

```python
kl_divergence(p=sunspots.Sunspots, q=forecast_sun.yhat.loc[sunspots.index])
```
